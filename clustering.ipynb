#Clustering

1.What is unsupervised learning in the context of machine learning?

ans-

    Unsupervised learning is a type of machine learning where the model learns patterns from unlabeled data without predefined outputs. Clustering is a common unsupervised task.

2.How does K-Means clustering algorithm work?

ans-

    K-means clustering algorithm works as-
    
    It chooses K initial centroids

    Assigns points to nearest centroid

    Recalculates centroids

    Repeats until centroids stabilize

3.Explain the concept of a dendrogram in hierarchical clustering    

ans-

    A dendrogram is a tree-like diagram that shows how data points merge into clusters at different distance levels in hierarchical clustering.

4.What is the main difference between K-Means and Hierarchical Clustering?    

ans-

    
    K-Means	              Hierarchical
    Needs K beforehand	  No need to specify K
    Fast & scalable	      Computationally expensive
    Spherical clusters	  Flexible shapes

5. What are the advantages of DBSCAN over K-Means?

ans-

    The advantages of DBSCAN over k-means-

    Detects arbitrary-shaped clusters

    Identifies noise/outliers

    No need to predefine number of clusters

6.When would you use Silhouette score in clustering?

ans-

    Used to evaluate clustering quality by measuring how similar a point is to its cluster vs other clusters.

7.What are the limitations of Hierarchical Clustering?

ans-

    High memory usage

    Not scalable

    Once merged/split, cannot undo    

8.Why is feature scaling important in clustering algorithms like K-Means?

ans-

    K-Means uses Euclidean distance, so features with larger scales dominate results.

9.How does DBSCAN identify noise points?

ans-

    Points that do not have enough neighbors within ε radius are labeled as noise.

10.Define inertia in context of K-Means.

ans-

    Inertia is the sum of squared distances between points and their assigned cluster centroids.

11.What is the elbow method in k-means clustering ?

ans-

    Plots inertia vs K. The elbow point indicates optimal number of clusters.

12.Describe the concept of "density" in DBSCAN.

ans-

    Density = number of points within ε-radius. Dense areas form clusters.

13.Can hierarchical clustering be used on categorical data?

ans-

    Yes, with appropriate distance measures (e.g., Hamming distance).

14.What does a negative Silhouette Score indicate?

ans-

    The point is likely assigned to the wrong cluster.

15.Explain the term "linkage criteria" in hierarchical clustering.

ans-

    Defines how distance between clusters is calculated (single, complete, average).

16.Why might K-Means clustering perform poorly on data with varying cluster sizes or densities?

ans-

    It assumes equal-sized spherical clusters, which fails for uneven densities.

17.What are the core parameters in DBSCAN, and how do they influence clustering?

ans-

    eps: neighborhood radius

    min_samples: minimum points to form cluster

18.How does K-Means++ improve upon standard K-Means initialization?

ans-

    Selects initial centroids far apart, reducing poor convergence.

19.What is agglomerative clustering?

ans-

    Bottom-up hierarchical clustering where each point starts as its own cluster.

20.What makes Silhouette Score a better metric than just inertia for model evaluation?

ans-

    Inertia only measures compactness; silhouette also measures separation.

21.Generate synthetic data with 4 centers using make_blobs and apply K-Means clustering. Visualize using a
scatter plot

```python
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

X, _ = make_blobs(n_samples=300, centers=4, random_state=42)
kmeans = KMeans(n_clusters=4)
y = kmeans.fit_predict(X)

plt.scatter(X[:,0], X[:,1], c=y)
plt.scatter(kmeans.cluster_centers_[:,0], kmeans.cluster_centers_[:,1], s=200)
plt.show()
```

32.Load the Digits dataset and apply DBSCAN after reducing dimensions with t-SNE. Visualize the results

```python
from sklearn.datasets import load_digits
from sklearn.manifold import TSNE
from sklearn.cluster import DBSCAN
import matplotlib.pyplot as plt

# Load Digits dataset
digits = load_digits()
X = digits.data

# Reduce dimensions using t-SNE
X_tsne = TSNE(n_components=2, random_state=42).fit_transform(X)

# Apply DBSCAN
dbscan = DBSCAN(eps=5, min_samples=5)
labels = dbscan.fit_predict(X_tsne)

# Visualize the clustering result
plt.figure(figsize=(8,6))
plt.scatter(X_tsne[:, 0], X_tsne[:, 1], c=labels)
plt.title("DBSCAN Clustering on Digits Dataset after t-SNE")
plt.xlabel("t-SNE Component 1")
plt.ylabel("t-SNE Component 2")
plt.show()
```

34. Load the Breast Cancer dataset and compare inertia values for K = 2 to 6 using K-Means. Show results in a
line plot

```python
from sklearn.datasets import load_breast_cancer
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt

# Load Breast Cancer dataset
data = load_breast_cancer()
X = data.data

# Scale the features
X_scaled = StandardScaler().fit_transform(X)

# Calculate inertia for K = 2 to 6
k_values = range(2, 7)
inertia_values = []

for k in k_values:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(X_scaled)
    inertia_values.append(kmeans.inertia_)

# Plot inertia values
plt.figure()
plt.plot(k_values, inertia_values, marker='o')
plt.xlabel("Number of Clusters (K)")
plt.ylabel("Inertia")
plt.title("Inertia vs K for Breast Cancer Dataset")
plt.show()
```

36.Use the Wine dataset, apply DBSCAN after scaling the data, and count the number of clusters (excluding
noise6

```python
from sklearn.datasets import load_wine
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import DBSCAN
import numpy as np

# Load Wine dataset
data = load_wine()
X = data.data

# Scale the data
X_scaled = StandardScaler().fit_transform(X)

# Apply DBSCAN
dbscan = DBSCAN(eps=1.5, min_samples=5)
labels = dbscan.fit_predict(X_scaled)

# Count clusters excluding noise (-1)
clusters = set(labels)
num_clusters = len(clusters) - (1 if -1 in clusters else 0)

print("Number of clusters (excluding noise):", num_clusters)
```

37.Generate synthetic data with make_blobs and apply KMeans. Then plot the cluster centers on top of the

38.Load the Iris dataset, cluster with DBSCAN, and print how many samples were identified as noise

```python
from sklearn.datasets import load_iris
from sklearn.cluster import DBSCAN
import numpy as np

# Load the Iris dataset
iris = load_iris()
X = iris.data

# Apply DBSCAN. The parameters eps and min_samples might need tuning for optimal results.
dbscan = DBSCAN(eps=0.5, min_samples=5)
labels = dbscan.fit_predict(X)

# Count the number of noise samples (labeled as -1 by DBSCAN)
noise_samples = np.sum(labels == -1)

print(f"Number of samples identified as noise: {noise_samples}")
```

46.Load the Wine dataset, apply KMeans, and visualize the cluster assignments in a seaborn pairplot (first 4
features)

```python
from sklearn.datasets import load_iris
from sklearn.cluster import KMeans
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the Iris dataset
iris = load_iris()
X = iris.data
y = iris.target # True labels (for reference, not used in clustering)
feature_names = iris.feature_names

# Apply K-Means clustering (assuming 3 clusters for Iris dataset)
kmeans = KMeans(n_clusters=3, random_state=42, n_init=10)
cluster_labels = kmeans.fit_predict(X)

# Create a DataFrame for visualization
# Using only the first 4 features as requested
df = pd.DataFrame(X[:, :4], columns=feature_names[:4])
df['Cluster'] = cluster_labels

# Visualize the cluster assignments using a seaborn pairplot
sns.pairplot(df, hue='Cluster', palette='viridis')
plt.suptitle('K-Means Clustering on Iris Dataset (first 4 features)', y=1.02)
plt.show()
```
